```python
import pandas as pd

df = pd.read_csv("https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD")
```

```python
# Getting all the cars and their types that do not qualify for clean alternative fuel vehicle

# Filtering out the data which contains not eligible
Filter = df[df["Clean Alternative Fuel Vehicle (CAFV) Eligibility"]=='Not eligible due to low battery range' ]

#Filtering out all the cars and types that don't qualify for CAFV
# note:- VIN = Vehicle identification number
ineligible_cars = Filter[['VIN (1-10)','Electric Vehicle Type']]

#returning the filtered out data accordingly
ineligible_cars
```

```python
#Getting all TESLA cars with the model year, and model type made in Bothell City.

# Filtering out the data which contains 'Tesla' and made in 'Bothell' city
Filter = df[(df['City'] == 'Bothell') & (df['Make'] == 'TESLA')]

#Filtering out all the Tesla made in Bothell city their model year and type.
tesla_cars = Filter[['Model Year', 'Make', 'Electric Vehicle Type']]

tesla_cars
```

```python
#Getting all the cars that have an electric range of more than 100, and were made after 2015 

#filtering out the data where electric range is more than 100 and were made after 2015
Filter = df[(df['Electric Range']>100) & (df['Model Year']>2015)]

range_abv_100 = Filter[['VIN (1-10)', 'Electric Range', 'Model Year']]

range_abv_100
```

```python
#Draw plots to show the distribution between city and electric vehicle type
import seaborn as sns
import matplotlib.pyplot as plt


# Drawing a count plot for visualisation of distribution
plt.figure(figsize=(100, 15))
sns.countplot(data=df, x='City', hue='Electric Vehicle Type')
plt.title('Distribution of Electric Vehicle Type by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Electric Vehicle Type')
plt.show()
# make sure to double click zoom for the better visbility of visualisation
```
